```python
import redis,yaml,uuid,time,datetime,ipaddress
import kubernetes
import warnings
warnings.filterwarnings('ignore')

#config
looptime=2
namespace_prefix='fah'
hostname_prefix='freeaihub'
pooltypes=[['v','create_v_pool_deploy',5],['c','create_c_pool_deploy',5],['k','create_k_pool_deploy',5],['k8sv','create_k8sv_pool_deploy',5]]


APISERVER = 'https://192.168.1.10:6443'
Token = open('token','r').read().strip('\n')
#pooll pool list
#poolh pool hash
#publ public list
#pubh public hash



r = redis.StrictRedis(host='localhost', port=6379, db=0)

# config.load_kube_config()
# Create a configuration object
configuration = kubernetes.client.Configuration()

# Specify the endpoint of your Kube cluster
configuration.host = APISERVER

# Security part.
# In this simple example we are not going to verify the SSL certificate of
# the remote cluster (for simplicity reason)
configuration.verify_ssl = False

# Nevertheless if you want to do it you can with these 2 parameters
# configuration.verify_ssl=True
# ssl_ca_cert is the filepath to the file that contains the certificate.
# configuration.ssl_ca_cert="certificate"
configuration.api_key = {"authorization": "Bearer " + Token}

# configuration.api_key["authorization"] = "bearer " + Token
# configuration.api_key_prefix['authorization'] = 'Bearer'
# configuration.ssl_ca_cert = 'ca.crt'
# Create a ApiClient with our config
kubernetes.client.Configuration.set_default(configuration)
# v1 API
v1 = kubernetes.client.CoreV1Api()

# CustomObjects API
k8s_CustomObjectsApi = kubernetes.client.CustomObjectsApi(kubernetes.client.ApiClient(configuration))
group = 'kubeovn.io' # str | The custom resource's group name
version = 'v1' # str | The custom resource's version
plural = 'subnets' # str | The custom resource's plural name. For TPRs this would be lowercase plural kind.

# Networking API
k8s_NetworkingV1Api = kubernetes.client.NetworkingV1Api(kubernetes.client.ApiClient(configuration))


def get_time():
    return time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())

def dateToStr(date):
    DATETIME_FMT = '%Y-%m-%d %H:%M:%S'
    return datetime.datetime.strftime(date, DATETIME_FMT)

def strToDate(strdate):
    DATETIME_FMT = '%Y-%m-%d %H:%M:%S'
    return datetime.datetime.strptime(strdate, DATETIME_FMT) 

def get_newid():
    return str(uuid.uuid4())[:8]


def showallkeys():
    print(r.keys())
    print('list namespace------')
    for ns in r.smembers("ns"):
        print(ns)
        
    print('list pool-list------')
    print(r.lrange('pooll',0,-1))
    print('list pool-hash------')
    print(r.hgetall('poolh'))
    
    print('list pub-list------')
    print(r.lrange('publ',0,-1))
    print('list pub-hash------')
    print(r.hgetall('pubh'))

def list_pool():
    for pod in r.lrange('publ',0,-1):
        print(pod)

#Template
def get_ns(nsid):
    template='''
apiVersion: v1
kind: Namespace
metadata:
  name: {namespace_prefix}{nsid}
'''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_subnet():
    return r.spop('subnet')

def get_subnets(nsid,subnet):
    template='''
apiVersion: kubeovn.io/v1
kind: Subnet
metadata:
  name: {namespace_prefix}{nsid}
spec:
  private: false
  natOutgoing: true
  protocol: IPv4
  cidrBlock: {subnet}.0/24
  excludeIps:
  - {subnet}.1..{subnet}.10
  gateway: {subnet}.1
  namespaces:
  - {namespace_prefix}{nsid}
'''
    subnet=subnet.split('.')[0]+'.'+subnet.split('.')[1]+'.'+subnet.split('.')[2]
    template=template.replace('{nsid}',nsid)
    template=template.replace('{subnet}',subnet)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_nspolicy(nsid,subnet):
    template='''
kind: NetworkPolicy
apiVersion: networking.k8s.io/v1
metadata:
    name: "{namespace_prefix}{nsid}"
    namespace: '{namespace_prefix}{nsid}'
spec:
#   policyTypes:
#   - Ingress
#   - Egress
    podSelector:
        matchLabels:
            ns: {namespace_prefix}{nsid}
    ingress:
        - from:
            - ipBlock:
                cidr: 10.16.0.0/16  #allow the pods that under the default namespace
            - podSelector:
                matchLabels:
                   ns: {namespace_prefix}{nsid}
    egress:
        - to:
            - ipBlock:
                cidr: 0.0.0.0/0
                except:
                    - 192.168.1.0/24
            # ports:
            # - protocol: TCP
            #   port: 22
'''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{subnet}',subnet.split('.')[0]+'.'+subnet.split('.')[1]+'.'+subnet.split('.')[2])
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_vmpod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
  annotations:
    kubernetes.io/target-runtime: virtlet.cloud
    VirtletCloudInitUserData: |
      ssh_pwauth: True
      runcmd:
      - echo 'Welcome to FreeAIHub Service!' > /etc/motd
      - /init.sh {token}
    # set root volume size
    #VirtletRootVolumeSize: 4Gi
    VirtletVCPUCount: "1"
spec:
  dnsPolicy: "None"
  dnsConfig:
    nameservers:
      - 223.5.5.5
    searches:
      - svc.cluster.local
      - cluster.local
    options:
      - name: ndots
        value: "2"
      - name: edns0
  # nodeSelector:
  #  extraRuntime: virtlet
  terminationGracePeriodSeconds: 120
  containers:
  - name: vm
    image: virtlet.cloud/vm
    imagePullPolicy: IfNotPresent
    tty: true
    stdin: true
    resources:
      limits:
        memory: 2Gi
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_k8svpod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
  annotations:
    kubernetes.io/target-runtime: virtlet.cloud
    VirtletCloudInitUserData: |
      ssh_pwauth: True
      runcmd:
      - echo 'Welcome to FreeAIHub Service!' > /etc/motd
      - /init.sh {token}
    # set root volume size
    #VirtletRootVolumeSize: 4Gi
    VirtletVCPUCount: "2"
spec:
  dnsPolicy: "None"
  dnsConfig:
    nameservers:
      - 223.5.5.5
    searches:
      - svc.cluster.local
      - cluster.local
    options:
      - name: ndots
        value: "2"
      - name: edns0
  # nodeSelector:
  #  extraRuntime: virtlet
  terminationGracePeriodSeconds: 120
  containers:
  - name: vm
    image: virtlet.cloud/k8sv
    imagePullPolicy: IfNotPresent
    tty: true
    stdin: true
    resources:
      limits:
        memory: 2Gi
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template


def get_katapod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  annotations:
    io.kubernetes.cri.untrusted-workload: "true"
    ovn.kubernetes.io/ingress_rate: "8"     #8MB
    ovn.kubernetes.io/egress_rate: "8"      #8MB
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
spec:
  # nodeSelector: 
  #   name: master      
  containers:
  - name: kata-container
    image: pod:v1
    imagePullPolicy: IfNotPresent
    #command: ["/usr/bin/supervisord"]
    command: ["bash","-c","/init.sh"]
    #command: ["bash","-c",'--','while true; do sleep 20; done;']
    ports:
      - containerPort: 22
      - containerPort: 88
      - containerPort: 5900   
      - containerPort: 6080
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
    env:
    - name: TOKEN
      value: "{token}"
    resources:
      limits:
        cpu: 500m
        memory: 2Gi  
  restartPolicy: Always
  dnsPolicy: None
  dnsConfig:
    nameservers:       
      - 223.5.5.5
    searches:
      - svc.cluster.local
      - cluster.local
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template


def get_svc_ssh(nsid,podid,svcname='ssh'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 22
    protocol: TCP
    targetPort: 22
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_cell(nsid,podid,svcname='cell'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: 8888
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_novnc(nsid,podid,svcname='novnc'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: 6080
  selector:
    name: {podid}
  type: ClusterIP 
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_http(nsid,podid,svcname='http',targetPort='80'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: {targetPort}
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    template=template.replace('{targetPort}',targetPort)
    return template

#operation
def check_pool():
    for pooltype in pooltypes:
        backtype = pooltype[0]
        poolmax = pooltype[-1]
        current_pool_num = r.llen(backtype+'-pooll')
        if current_pool_num < poolmax:
            for i in range(poolmax - current_pool_num):
                nsid = get_newid()
                nsid = backtype+nsid#不同类别前缀标识
                namespace=namespace_prefix+nsid
                subnet= bytes.decode(get_subnet())
                token = backtype+get_newid()
                
                v1.create_namespace(yaml.safe_load(get_ns(nsid)))
                eval(pooltype[1])(backtype,namespace,nsid,subnet,token)
                
                r.hset(backtype+'-poolh',namespace_prefix+nsid+'-token',token)#token
                r.hset(backtype+'-poolh',namespace_prefix+nsid+'-subnet',subnet)#subnet
                
                r.rpush(backtype+'-pooll',namespace_prefix+nsid)
                print(nsid,'type a deploy created!')
                
                
def create_v_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single vm in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix)))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,hostname_prefix,token)))
    
    
def create_c_pool_deploy(backtype,namespace,nsid,subnet,token):
    # backend c type --- 2 node in cluster,master and slave
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,'master')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,'master',)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,'master',)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,'master')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,'master',token)))
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,'node1')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,'node1',token)))

def create_k_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single kata container in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix)))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_katapod(nsid,hostname_prefix,token)))
    

def create_k8sv_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single vm in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix,targetPort='8080')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_k8svpod(nsid,hostname_prefix,token)))
    
def del_pool(nsid):
    v1.delete_namespace(nsid)
    k8s_CustomObjectsApi.delete_cluster_custom_object(group, version, plural,nsid)


def delete_expire_pub():
    print('loop the delete_expire_pub')
    for pooltype in pooltypes:
        envtype=pooltype[0]
        #对各自类别中的pub当前列表进行遍历检查
        for i,envid in enumerate(r.lrange(envtype+'-publ',0,-1)):
            envid=bytes.decode(envid)
            print('  delete_expire_pub check all pod in pub')
            if r.hexists(envtype+'-pubh',envid+'-id') and (datetime.datetime.now() > strToDate(bytes.decode(r.hget(envtype+'-pubh',envid+"-lifeend")))):
              
                #删除publ列表中的key
                r.lrem(envtype+'-publ',i-1,envid)#删publ中key
                print(envtype,'=====',envid)
                
                #删除对应k8s,namespace and subnet
                del_pool(bytes.decode(r.hget(envtype+'-pubh',envid+'-id')))#删ns,subnet
                
                #删除pubh中所有的相关字段 
                envclientid = bytes.decode(r.hget(envtype+'-pubh',envid))#envclientid
                r.hdel(envtype+'-pubh',envclientid)
                r.hdel(envtype+'-pubh',envid)
                r.hdel(envtype+'-pubh',envid+'-id')
                r.hdel(envtype+'-pubh',envid+'-lifeend')

                print('    delete_expire_pub deleted:',envid)


def get_allns(ns='ns'):
    api_response = v1.list_namespace()
    print('get_allns')
    for item in api_response.items:
        nsname=item.metadata.name
        if nsname in ['default','kube-ovn','kube-public','kube-system','kube-node-lease']:continue
        r.sadd('ns',nsname)
        
def get_serviceip():
        #for ns in r.lrange('ns',0,-1):
        print('get_all service in ns')
        for ns in r.smembers("ns"):
            namespace=bytes.decode(ns)
            ret = v1.list_namespaced_service(namespace,watch=False)
            for serviceitem in ret.items:
                servicename=serviceitem.metadata.name
                serviceip=serviceitem.spec.cluster_ip
    #             print(namespace,servicename,serviceip)
                r.hset(namespace,servicename,serviceip)

#INIT
def clear_allkeys():
    for item in (r.keys()):
        r.delete(item)
        
def init_subnet():
    for i1 in range(20,200):
        for i2 in range(20,200):
            subnet='172.'+str(i1)+'.'+str(i2)+'.0/24'
            r.sadd('subnet',subnet)
#             ipaddress.ip_network(subnet)
            
def init_namespace():
    #清除redis所有key
    clear_allkeys()
    
    init_subnet()
    
    #list namespace and delete fah-*
    for item in v1.list_namespace().items:
        nsname=item.metadata.name
        print(nsname)
        if nsname.startswith(namespace_prefix):
            print('ns: ',nsname,'deleted')
            v1.delete_namespace(nsname)
            
    #list subnet and delete fah-*
    for item in k8s_CustomObjectsApi.list_cluster_custom_object(group, version, plural)['items']:
        subnetname=item['metadata']['name']
        print(subnetname)
        if subnetname.startswith(namespace_prefix):
            print('subnet: ',subnetname,' deleted')
            k8s_CustomObjectsApi.delete_cluster_custom_object(group, version, plural, subnetname)
            
            
init_namespace()
time.sleep(10)
while True:
    get_allns()
    get_serviceip()
    check_pool()
    time.sleep(looptime)
    delete_expire_pub()
```

```python
import redis,yaml,uuid,time,datetime,ipaddress
import kubernetes
import warnings
warnings.filterwarnings('ignore')

#config
looptime=2
namespace_prefix='fah'
hostname_prefix='freeaihub'
pooltypes=[['v','create_v_pool_deploy',5],['c','create_c_pool_deploy',5],['k','create_k_pool_deploy',5],['k8sv','create_k8sv_pool_deploy',5]]


APISERVER = 'https://192.168.1.10:6443'
Token = open('token','r').read().strip('\n')
#pooll pool list
#poolh pool hash
#publ public list
#pubh public hash



r = redis.StrictRedis(host='localhost', port=6379, db=0)

# config.load_kube_config()
# Create a configuration object
configuration = kubernetes.client.Configuration()

# Specify the endpoint of your Kube cluster
configuration.host = APISERVER

# Security part.
# In this simple example we are not going to verify the SSL certificate of
# the remote cluster (for simplicity reason)
configuration.verify_ssl = False

# Nevertheless if you want to do it you can with these 2 parameters
# configuration.verify_ssl=True
# ssl_ca_cert is the filepath to the file that contains the certificate.
# configuration.ssl_ca_cert="certificate"
configuration.api_key = {"authorization": "Bearer " + Token}

# configuration.api_key["authorization"] = "bearer " + Token
# configuration.api_key_prefix['authorization'] = 'Bearer'
# configuration.ssl_ca_cert = 'ca.crt'
# Create a ApiClient with our config
kubernetes.client.Configuration.set_default(configuration)
# v1 API
v1 = kubernetes.client.CoreV1Api()

# CustomObjects API
k8s_CustomObjectsApi = kubernetes.client.CustomObjectsApi(kubernetes.client.ApiClient(configuration))
group = 'kubeovn.io' # str | The custom resource's group name
version = 'v1' # str | The custom resource's version
plural = 'subnets' # str | The custom resource's plural name. For TPRs this would be lowercase plural kind.

# Networking API
k8s_NetworkingV1Api = kubernetes.client.NetworkingV1Api(kubernetes.client.ApiClient(configuration))


def get_time():
    return time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())

def dateToStr(date):
    DATETIME_FMT = '%Y-%m-%d %H:%M:%S'
    return datetime.datetime.strftime(date, DATETIME_FMT)

def strToDate(strdate):
    DATETIME_FMT = '%Y-%m-%d %H:%M:%S'
    return datetime.datetime.strptime(strdate, DATETIME_FMT) 

def get_newid():
    return str(uuid.uuid4())[:8]


def showallkeys():
    print(r.keys())
    print('list namespace------')
    for ns in r.smembers("ns"):
        print(ns)
        
    print('list pool-list------')
    print(r.lrange('pooll',0,-1))
    print('list pool-hash------')
    print(r.hgetall('poolh'))
    
    print('list pub-list------')
    print(r.lrange('publ',0,-1))
    print('list pub-hash------')
    print(r.hgetall('pubh'))

def list_pool():
    for pod in r.lrange('publ',0,-1):
        print(pod)

#Template
def get_ns(nsid):
    template='''
apiVersion: v1
kind: Namespace
metadata:
  name: {namespace_prefix}{nsid}
'''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_subnet():
    return r.spop('subnet')

def get_subnets(nsid,subnet):
    template='''
apiVersion: kubeovn.io/v1
kind: Subnet
metadata:
  name: {namespace_prefix}{nsid}
spec:
  private: false
  natOutgoing: true
  protocol: IPv4
  cidrBlock: {subnet}.0/24
  excludeIps:
  - {subnet}.1..{subnet}.10
  gateway: {subnet}.1
  namespaces:
  - {namespace_prefix}{nsid}
'''
    subnet=subnet.split('.')[0]+'.'+subnet.split('.')[1]+'.'+subnet.split('.')[2]
    template=template.replace('{nsid}',nsid)
    template=template.replace('{subnet}',subnet)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_nspolicy(nsid,subnet):
    template='''
kind: NetworkPolicy
apiVersion: networking.k8s.io/v1
metadata:
    name: "{namespace_prefix}{nsid}"
    namespace: '{namespace_prefix}{nsid}'
spec:
#   policyTypes:
#   - Ingress
#   - Egress
    podSelector:
        matchLabels:
            ns: {namespace_prefix}{nsid}
    ingress:
        - from:
            - ipBlock:
                cidr: 10.16.0.0/16  #allow the pods that under the default namespace
            - podSelector:
                matchLabels:
                   ns: {namespace_prefix}{nsid}
    egress:
        - to:
            - ipBlock:
                cidr: 0.0.0.0/0
                except:
                    - 192.168.1.0/24
            # ports:
            # - protocol: TCP
            #   port: 22
'''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{subnet}',subnet.split('.')[0]+'.'+subnet.split('.')[1]+'.'+subnet.split('.')[2])
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_vmpod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
  annotations:
    kubernetes.io/target-runtime: virtlet.cloud
    VirtletCloudInitUserData: |
      ssh_pwauth: True
      runcmd:
      - echo 'Welcome to FreeAIHub Service!' > /etc/motd
      - /init.sh {token}
    # set root volume size
    #VirtletRootVolumeSize: 4Gi
    VirtletVCPUCount: "1"
spec:
  # nodeSelector:
  #  extraRuntime: virtlet
  terminationGracePeriodSeconds: 120
  containers:
  - name: vm
    image: virtlet.cloud/vm
    imagePullPolicy: IfNotPresent
    tty: true
    stdin: true
    resources:
      limits:
        memory: 2Gi
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template

def get_k8svpod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
  annotations:
    kubernetes.io/target-runtime: virtlet.cloud
    VirtletCloudInitUserData: |
      ssh_pwauth: True
      runcmd:
      - echo 'Welcome to FreeAIHub Service!' > /etc/motd
      - /init.sh {token}
    # set root volume size
    #VirtletRootVolumeSize: 4Gi
    VirtletVCPUCount: "2"
spec:
  dnsPolicy: "None"
  dnsConfig:
    nameservers:
      - 114.114.114.114
    searches:
      - svc.cluster.local
      - cluster.local
    options:
      - name: ndots
        value: "2"
      - name: edns0
  # nodeSelector:
  #  extraRuntime: virtlet
  terminationGracePeriodSeconds: 120
  containers:
  - name: vm
    image: virtlet.cloud/k8sv
    imagePullPolicy: IfNotPresent
    tty: true
    stdin: true
    resources:
      limits:
        memory: 2Gi
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template


def get_katapod(nsid,podid,token):
    template='''
apiVersion: v1
kind: Pod
metadata:
  name: {podid}
  namespace: {namespace_prefix}{nsid}
  annotations:
    io.kubernetes.cri.untrusted-workload: "true"
    ovn.kubernetes.io/ingress_rate: "8"     #8MB
    ovn.kubernetes.io/egress_rate: "8"      #8MB
  labels:
    name: {podid}
    ns: {namespace_prefix}{nsid}
spec:
  # nodeSelector: 
  #   name: master      
  containers:
  - name: kata-container
    image: pod:v1
    imagePullPolicy: IfNotPresent
    #command: ["/usr/bin/supervisord"]
    command: ["bash","-c","/init.sh"]
    #command: ["bash","-c",'--','while true; do sleep 20; done;']
    ports:
      - containerPort: 22
      - containerPort: 88
      - containerPort: 5900   
      - containerPort: 6080
    volumeMounts:
    - mountPath: /share
      name: share
      readOnly: false
    env:
    - name: TOKEN
      value: "{token}"
    resources:
      limits:
        cpu: 500m
        memory: 2Gi  
  restartPolicy: Always
  dnsPolicy: None
  dnsConfig:
    nameservers:       
      - 223.5.5.5
    searches:
      - svc.cluster.local
      - cluster.local
  volumes:
  - hostPath:
      path: /nfs
      type: ""
    name: share
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid)
    template=template.replace('{token}',token)
    template=template.replace('{namespace_prefix}',namespace_prefix)
    return template


def get_svc_ssh(nsid,podid,svcname='ssh'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 22
    protocol: TCP
    targetPort: 22
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_cell(nsid,podid,svcname='cell'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: 8888
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_novnc(nsid,podid,svcname='novnc'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: 6080
  selector:
    name: {podid}
  type: ClusterIP 
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    return template

def get_svc_http(nsid,podid,svcname='http',targetPort='80'):
    template='''
apiVersion: v1
kind: Service
metadata:
  labels:
    name: {podid}-{svcname}
  name: "{podid}-{svcname}"
  namespace: {namespace_prefix}{nsid}
spec:
  ports:
  - port: 80
    protocol: TCP
    targetPort: {targetPort}
  selector:
    name: {podid}
  type: ClusterIP  
    '''
    template=template.replace('{nsid}',nsid)
    template=template.replace('{podid}',podid.lower())
    template=template.replace('{namespace_prefix}',namespace_prefix)
    template=template.replace('{svcname}',svcname)
    template=template.replace('{targetPort}',targetPort)
    return template

#operation
def check_pool():
    for pooltype in pooltypes:
        backtype = pooltype[0]
        poolmax = pooltype[-1]
        current_pool_num = r.llen(backtype+'-pooll')
        if current_pool_num < poolmax:
            for i in range(poolmax - current_pool_num):
                nsid = get_newid()
                nsid = backtype+nsid#不同类别前缀标识
                namespace=namespace_prefix+nsid
                subnet= bytes.decode(get_subnet())
                token = backtype+get_newid()
                
                v1.create_namespace(yaml.safe_load(get_ns(nsid)))
                eval(pooltype[1])(backtype,namespace,nsid,subnet,token)
                
                r.hset(backtype+'-poolh',namespace_prefix+nsid+'-token',token)#token
                r.hset(backtype+'-poolh',namespace_prefix+nsid+'-subnet',subnet)#subnet
                
                r.rpush(backtype+'-pooll',namespace_prefix+nsid)
                print(nsid,'type a deploy created!')
                
                
def create_v_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single vm in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix)))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,hostname_prefix,token)))
    
    
def create_c_pool_deploy(backtype,namespace,nsid,subnet,token):
    # backend c type --- 2 node in cluster,master and slave
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,'master')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,'master',)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,'master',)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,'master')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,'master',token)))
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,'node1')))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,'node1')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_vmpod(nsid,'node1',token)))

def create_k_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single kata container in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix)))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_katapod(nsid,hostname_prefix,token)))
    

def create_k8sv_pool_deploy(backtype,namespace,nsid,subnet,token):
    #default type backend --- single vm in subnet
    k8s_CustomObjectsApi.create_cluster_custom_object(group, version, plural, yaml.safe_load(get_subnets(nsid,subnet)))
    k8s_NetworkingV1Api.create_namespaced_network_policy(namespace, yaml.safe_load(get_nspolicy(nsid,subnet)))
    time.sleep(looptime)
    
    v1.create_namespaced_service(namespace, yaml.safe_load(get_svc_ssh(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_novnc(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_cell(nsid,hostname_prefix)))
    v1.create_namespaced_service(namespace,yaml.safe_load(get_svc_http(nsid,hostname_prefix,targetPort='8080')))
    v1.create_namespaced_pod(namespace, yaml.safe_load(get_k8svpod(nsid,hostname_prefix,token)))
    
def del_pool(nsid):
    v1.delete_namespace(nsid)
    k8s_CustomObjectsApi.delete_cluster_custom_object(group, version, plural,nsid)


def delete_expire_pub():
    print('loop the delete_expire_pub')
    for pooltype in pooltypes:
        envtype=pooltype[0]
        #对各自类别中的pub当前列表进行遍历检查
        for i,envid in enumerate(r.lrange(envtype+'-publ',0,-1)):
            envid=bytes.decode(envid)
            print('  delete_expire_pub check all pod in pub')
            if r.hexists(envtype+'-pubh',envid+'-id') and (datetime.datetime.now() > strToDate(bytes.decode(r.hget(envtype+'-pubh',envid+"-lifeend")))):
              
                #删除publ列表中的key
                r.lrem(envtype+'-publ',i-1,envid)#删publ中key
                print(envtype,'=====',envid)
                
                #删除对应k8s,namespace and subnet
                del_pool(bytes.decode(r.hget(envtype+'-pubh',envid+'-id')))#删ns,subnet
                
                #删除pubh中所有的相关字段 
                envclientid = bytes.decode(r.hget(envtype+'-pubh',envid))#envclientid
                r.hdel(envtype+'-pubh',envclientid)
                r.hdel(envtype+'-pubh',envid)
                r.hdel(envtype+'-pubh',envid+'-id')
                r.hdel(envtype+'-pubh',envid+'-lifeend')

                print('    delete_expire_pub deleted:',envid)


def get_allns(ns='ns'):
    api_response = v1.list_namespace()
    print('get_allns')
    for item in api_response.items:
        nsname=item.metadata.name
        if nsname in ['default','kube-ovn','kube-public','kube-system','kube-node-lease']:continue
        r.sadd('ns',nsname)
        
def get_serviceip():
        #for ns in r.lrange('ns',0,-1):
        print('get_all service in ns')
        for ns in r.smembers("ns"):
            namespace=bytes.decode(ns)
            ret = v1.list_namespaced_service(namespace,watch=False)
            for serviceitem in ret.items:
                servicename=serviceitem.metadata.name
                serviceip=serviceitem.spec.cluster_ip
    #             print(namespace,servicename,serviceip)
                r.hset(namespace,servicename,serviceip)

#INIT
def clear_allkeys():
    for item in (r.keys()):
        r.delete(item)
        
def init_subnet():
    for i1 in range(20,200):
        for i2 in range(20,200):
            subnet='172.'+str(i1)+'.'+str(i2)+'.0/24'
            r.sadd('subnet',subnet)
#             ipaddress.ip_network(subnet)
            
def init_namespace():
    #清除redis所有key
    clear_allkeys()
    
    init_subnet()
    
    #list namespace and delete fah-*
    for item in v1.list_namespace().items:
        nsname=item.metadata.name
        print(nsname)
        if nsname.startswith(namespace_prefix):
            print('ns: ',nsname,'deleted')
            v1.delete_namespace(nsname)
            
    #list subnet and delete fah-*
    for item in k8s_CustomObjectsApi.list_cluster_custom_object(group, version, plural)['items']:
        subnetname=item['metadata']['name']
        print(subnetname)
        if subnetname.startswith(namespace_prefix):
            print('subnet: ',subnetname,' deleted')
            k8s_CustomObjectsApi.delete_cluster_custom_object(group, version, plural, subnetname)
            
            
init_namespace()
time.sleep(10)
while True:
    get_allns()
    get_serviceip()
    check_pool()
    time.sleep(looptime)
    delete_expire_pub()
```
